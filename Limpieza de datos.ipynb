```python
import numpy as np
import os.path
import pandas as pd
from statistics import mean, mode
from datetime import datetime, timedelta
import random
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
###########################################################
from tqdm import tqdm# barras de progreso
###########################################################
#Para no visualizar la salida de la intruccion to_datetime
from contextlib import redirect_stdout
import io

path = os.getcwd()
os.getcwd()

dataframe = pd.read_csv('C:\\Users\\PC\\Desktop\\Narco\\entropía\\base_actualizada.csv')
# Eliminar la información de la hora en las columnas de fecha
dataframe['date_start'] = dataframe['date_start'].str.split().str[0]
dataframe['date_end'] = dataframe['date_end'].str.split().str[0]
# Creamos un flujo nulo para redirigir la salida
null_stream = io.StringIO()
with redirect_stdout(null_stream):
    dataframe['date_start'] = pd.to_datetime(dataframe['date_start'], format='%m/%d/%Y')
    dataframe['date_end'] = pd.to_datetime(dataframe['date_end'], format='%m/%d/%Y')
#Calculamos la duración del conflicto
dataframe['duration'] = dataframe['date_end']-dataframe['date_start']
# Filtra los eventos cuya duración sea menor o igual a 3 días
filtered_df = dataframe[dataframe['duration'] <= pd.Timedelta(days=3)]
print("Número de columnas, filas:", dataframe.shape)
print("Número de columnas, filas:", filtered_df.shape)
database = pd.DataFrame()
database['date'] = filtered_df['date_start']
database['adm_1'] = filtered_df['adm_1']
database['adm_2'] = filtered_df['adm_2']
database['cart_a'] = filtered_df['side_a']
database['cart_b'] = filtered_df['side_b']
database['mue_a'] = filtered_df['deaths_a']
database['mue_b'] = filtered_df['deaths_b']
database['mue_civ'] = filtered_df['deaths_civilians']
database['mue_desc'] = filtered_df['deaths_unknown']
database['best_est'] = filtered_df['best_est']
database['latitude'] = filtered_df['latitude']
database['longitude'] = filtered_df['longitude']
# Ordenar el DataFrame por la columna de fecha en orden descendente
database = database.sort_values(by='date', ascending=True)
# Filtrar el DataFrame para eliminar las fechas menores a 2004
#PUES ENTRE 1989 Y 2003 SOLO HUBO 43 EVENTOS, A PARTIR DE 2004 EMPIEZA UNA MAYOR FRECUENCIA ANUAL
database = database[database['date'].dt.year >= 2004]
print("Número de columnas, filas:", database.shape)
# Eliminar las filas que contienen valores NaN en cualquier columna
database = database.dropna()
database = database.reset_index(drop=True)
database.to_csv('Actualizada_filtrada.csv', index=False)
```
